#### 1. Record Reader
Descreva uma função que realize a leitura de dados "mercado_ti.txt" e "data_science.txt"

```python
import pandas as pd

def recordReader(filename):
    data = open(filename).read()
    return data

mercado = recordReader('mercado_ti.txt')
data_science = recordReader('data_science.txt')
```

#### 2. Input split
defina uma função split que recebe os dados e quebre os em partes menores para usar como entrada na função mapper.
Exemplo: string.split()

```python
def input_split(data):
    return data.split(" ")

mercado_split = input_split(mercado)
data_science_split = input_split(data_science)
```

#### 3. Map
O usuário define uma função intermediária que recebe os dados e mapeia  pares de chave e valor (definido pelo usuário)

```python
def mapper(array):
#     {'palavra', chave}
    dic = []
    for word in array:
        dic.append((word, 1))
    return dic

mercado_dic = mapper(mercado_split)
ds_dic = mapper(data_science_split)
```

#### 4. Shuffle
Descreva uma função Shufle que ordena, filtra e agrupam os dados para serem usados como entrada na função reducer

```python
def shufle(array):
    dic = {}
#     {'de': []}
    for word, value in array:
        if word not in dic:
            dic[word] = []
        dic[word].append(1)
    return dic

mercado_shuf = shufle(mercado_dic)
ds_shuf = shufle(ds_dic)
```

#### 5. Reduce
Use uma função que agregue dados pelo valor (v) de acordo com a chave (k) para enviar pares de chave-valor para saída

```python
def reducer(dic):
    dic2 = {}
    for key in dic:
        dic2[key] = sum(dic[key])
    return dic2

mercado_reduce = reducer(mercado_shuf)
ds_reduce = reducer(ds_shuf)
mercado_reduce
```

#### 6.Output Format
Forneça os pares de chave-valor finais ordenados pela maior frequência de palavras contadas.

```python
import operator

def outputFormat(dic):
    return sorted(dic.items(), key=operator.itemgetter(1))

teste = outputFormat(mercado_reduce)
teste
```
