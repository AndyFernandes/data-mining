# KNN 

```python
# importando bibliotecas
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
from sklearn.neighbors import KNeighborsClassifier, DistanceMetric
from sklearn.metrics import accuracy_score
from sklearn.model_selection import KFold, cross_val_score, train_test_split

def euc_distance(linha1, linha2):
    return dist.pairwise(linha1, linha2)[0,0]
```

```python
file = pd.read_csv('abalone.csv', names = ["Sex", "Lenght", "Diameter", "Height", "Whole", "Shucked", "Viscera", "Shell", "Rings" ])

for feature in ["Sex"]:
    file[feature] = file[feature].astype('category')
    file[feature] = file[feature].cat.codes

# Iremos dividir o nosso conjunto de dados em:
# - treino, as nossas features, represetado pelo X
# - teste, target, representado pelo Y
X = file.loc[:, ["Sex", "Lenght", "Diameter", "Height", "Whole","Shucked", "Viscera", "Shell"]]
Y = file.loc[:, ["Rings"]]

# Separação do dataset
# Treino (80%) / Teste (20%)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y.values.flatten(),train_size = 0.8, test_size = 0.2, random_state = 42)

dist = DistanceMetric.get_metric('euclidean')

#instanciando o modelo de aprendizado com k = 3
knn = KNeighborsClassifier(n_neighbors = 3, metric='euclidean')

# treinando o modelo com os conjuntos de treino
knn.fit(X_train, Y_train)

# realizando a predicao
pred = knn.predict(X_test)

acc = accuracy_score(Y_test, pred)
print(acc)

k_fold = KFold(n_splits = 5)
scores = cross_val_score(knn, X_train, Y_train, cv = k_fold)
#print(scores)
media = np.mean(scores) * 100
print("Media: %s%%" %(media))
```
