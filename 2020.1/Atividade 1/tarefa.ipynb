### 1 - Quantas e quais as marcas de produtos contidas no dataset?

```python
print("Total de marcas: ", data['brand'].unique().shape)
list(data['brand'].unique())
```

### 2 - Quais as marcas que aparecem mais no dataset? Quais aparecem menos?

```python
brands_asc = data[['id', 'brand']].groupby('brand').count().sort_values(by='id').rename(columns={"id":"count"})
brands_desc = data[['id', 'brand']].groupby('brand').count().sort_values(by='id', ascending=False).rename(columns={"id":"count"})

brands_asc
```

```python
brands_desc
```

### 3 - Compute a média e desvio padrão da frequência das marcas de produtos

```python
mean = data.shape[0]/brands_asc.shape[0]
variance = sum(list(map(lambda x: (x[0]-mean)**2, brands_asc.values)))/brands_asc.shape[0]
print('Média: ', mean)
print('Variancia: ', variance)
```

### 4 - Quais os ingredientes que aparecem nos produtos? Mostre quantos são, os que mais aparecem e os que menos aparecem.

```python
import re
import collections
ingredients = list(map(lambda row: str(row).lower().replace(" ", ",").replace(".", "").replace("(", ",").replace(")", ",").replace("and", ",").replace('\'', '').replace('/', '').replace('*', '').split(','), data['features.value'].to_list()))
ingredients = pd.DataFrame({'ing': [ing  for row in ingredients for ing in row if ing != '']})

ingredients_count = {}
for ing in ingredients['ing']:
    if ing in ingredients_count:
        ingredients_count[ing] += 1
    else:
        ingredients_count[ing] = 1

print("Quantidade de ingredientes: ", len(ingredients['ing'].unique()))


ingredients_count_desc = {w: ingredients_count[w] for w in sorted(ingredients_count, key=ingredients_count.get, reverse=True)}
ingredients_count_asc = {w: ingredients_count[w] for w in sorted(ingredients_count, key=ingredients_count.get)}
                                                                 
ingredients_count_desc
```

```python
ingredients_count_asc
```

### 5 - Compute a média e desvio padrão da frequência dos ingredientes

```python
mean_ing = data.shape[0]/len(ingredients['ing'].unique())
variance_ing = sum(list(map(lambda x: (x[0]-mean)**2, brands_asc.values)))/brands_asc.shape[0]
print('Média: ', mean_ing)
print('Variancia: ', variance_ing)
```
